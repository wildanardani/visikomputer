```
import cv2 
import numpy as np 
from google.colab.patches import cv2_imshow
```

```
image1 = cv2.imread('jokowi.jpg')
```

```
img = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
```

```
ret, thresh1 = cv2.threshold(img, 50, 210, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 50, 210, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img, 50, 210, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img, 50, 210, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img, 50, 210, cv2.THRESH_TOZERO_INV)
```

```
cv2_imshow(image1)
cv2_imshow(thresh1)
cv2_imshow(thresh2)
cv2_imshow(thresh3)
cv2_imshow(thresh4)
cv2_imshow(thresh5)
```
